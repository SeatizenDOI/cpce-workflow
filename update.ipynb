```python
from pathlib import Path
import shutil
import pandas as pd

SESSION_FOLDER = Path("/home/bioeos/Documents/Bioeos/aina_dataset")
```

```python
if not SESSION_FOLDER.exists() or not SESSION_FOLDER.is_dir():
    raise NameError("Not folder", SESSION_FOLDER)




for session in SESSION_FOLDER.iterdir():

    if "SNORK" in session.name:
        shutil.move(session, Path(session.parent, session.name.replace("SNORK", 'UVC')))


    metadata_path = Path(session, "METADATA", "metadata.csv")

    if not metadata_path.exists() or not metadata_path.is_file():
        print(f"Metadata file not found for {metadata_path}")
    
    metadata_df = pd.read_csv(metadata_path)
    metadata_df["OriginalFileName"] = metadata_df["FileName"]
    metadata_df["FileName"] = metadata_df.apply(lambda x: f'{session.name}_{str(int(x.name)+1).rjust(4, '0')}.JPG', axis=1)
    metadata_df = metadata_df.rename(columns={"Datetime": "DateTime"})


    # cols = metadata_df.columns.tolist()
    # cols = cols[2:] + cols[:2]
    # metadata_df = metadata_df[cols]
    metadata_df.to_csv(metadata_path, index=False)
```

```python
labels = {}
for file in Path("dataset/clean_data/CPCE").iterdir():
    with open(file, 'r', encoding='latin1') as f:

        for row in f:
            if "Notes" not in row: continue
            row = row.replace("\n", '').split(",")[1].replace('"', '')
            if row == '': continue
            if row not in labels:
                labels[row] = 0
            labels[row] += 1
            if row in ['Mta']:
                print(file)

print(labels)
[print(key, labels[key]) for key in labels]
```

```python
codes = {}
isFooter = False
with open("amoros/CPCe_benthic_codes_41_Madagascar_MPAv4_40.txt", "r") as f:
    for i, row in enumerate(f):
        if i < 15: continue
        if "NOTES" in row:
            isFooter = True
        if isFooter: continue
        row = row.replace("\n", "").replace('"', '').split(',')
        codes[row[0]] = row[1]

sorted_labels = dict(sorted(labels.items(), key=lambda item: item[1]))

for label in sorted_labels:
    if label in ['CCA', 'Ctn']:
        print(label, codes[label], labels[label])
```
